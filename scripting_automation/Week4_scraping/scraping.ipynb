```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
```

```python
# Get HTML from some source

url = 'https://en.wikipedia.org/wiki/list_of_best_selling_video_games'
page = urlopen(url)
print(page)

#Get data from website
# data_root = "" #url
# lifesat = pd.read_csv(data_root)
# lifesat.head()

soup = BeautifulSoup( page, 'html.parser' )
game_table = soup.find( 'table', { 'class' : { 'wikitable' } } ) # Find specific table

game_tbody = game_table.find( 'tbody' ) # Table body
game_tr = game_tbody.find_all('tr') # Look for rows in the table

row_list = [] # use to store values

# create a for-loop to get all the td's
for tr in game_tr:
    table_row = tr.find_all('td')
    for td in table_row:
        contents = ''
        if (td.find('sup')): # specific to this page
            the_sup = td.find('sup')
            the_sup.decompose() # remove tag - you might use this in the assignment
        # specific to this page, yours might look different
        if(td.find('a')):
            contents = td.find('a').contents[0]
        elif(td.find('span')):
            contents = td.find('span').contents[0] # There might be more types of HTML tags
        else:
            contents = td.contents[0]
            
        if (contents != ''): # if something was found
            contents = str(contents)
            contents = contents.replace('\n', '')
            row_list.append(contents) # Add to list
            
    if len(row_list) > 0:
        print(row_list)
```
